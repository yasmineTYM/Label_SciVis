```python
## append two columns: LabelKeyword, Tag 

import pandas as pd

data = pd.read_csv('vispub.csv')
```

```python
total_length = len(data)
temp_k = []
temp_t = []
for i in range(total_length):
    temp_k.append('')
    temp_t.append('todo')
data['LabelKeyword'] = temp_k
data['Tag'] = temp_t
```

```python
data.head(3)
```

```python
data.to_csv('vispub.csv',index=False)
```
